# Pogađanje dobne skupine 18-30:

```python
### 18-30 pogada 18-30

df_1830_1830 = df_18_30.drop(kod3050, axis=1).drop(kod5080, axis=1).drop(kod5080_score, axis=1).drop(kod3050_score, axis=1)
df_1830_1830 = df_1830_1830.reset_index(drop=True)

for column in kod1830_score:
    df_1830_1830[column] = df_1830_1830[column].apply(lambda x: float(x.split(' / ')[0]))

df_1830_1830['Bodovi'] = 0
for column in kod1830_score:
    df_1830_1830['Bodovi'] = df_1830_1830[column] + df_1830_1830['Bodovi']
    
df_1830_1830['Točnost'] = df_1830_1830['Bodovi'] /len(kod1830) *100

### 18-30 pogada 30-50

df_1830_3050 = df_18_30.drop(kod1830, axis=1).drop(kod5080, axis=1).drop(kod5080_score, axis=1).drop(kod1830_score, axis=1)
df_1830_3050 = df_1830_3050.reset_index(drop=True)

for column in kod3050_score:
    df_1830_3050[column] = df_1830_3050[column].apply(lambda x: float(x.split(' / ')[0]))

df_1830_3050['Bodovi'] = 0
for column in kod3050_score:
    df_1830_3050['Bodovi'] = df_1830_3050[column] + df_1830_3050['Bodovi']
    
df_1830_3050['Točnost'] = df_1830_3050['Bodovi'] /len(kod3050) *100

### 18-30 pogada 50-80

df_1830_5080 = df_18_30.drop(kod3050, axis=1).drop(kod1830, axis=1).drop(kod1830_score, axis=1).drop(kod3050_score, axis=1)
df_1830_5080 = df_1830_5080.reset_index(drop=True)

for column in kod5080_score:
    df_1830_5080[column] = df_1830_5080[column].apply(lambda x: float(x.split(' / ')[0]))

df_1830_5080['Bodovi'] = 0
for column in kod5080_score:
    df_1830_5080['Bodovi'] = df_1830_5080[column] + df_1830_5080['Bodovi']

df_1830_5080['Točnost'] = df_1830_5080['Bodovi'] /len(kod5080) *100
```
